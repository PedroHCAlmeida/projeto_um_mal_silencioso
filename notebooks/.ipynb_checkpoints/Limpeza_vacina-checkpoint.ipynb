```python
import pandas as pd
```

```python
vacinas = pd.read_csv('../dados_brutos/vacina_hepatite.csv', sep=';')
```

```python
vacinas
```

```python
vacinas = vacinas.drop(columns=[' Total'], index=[0,1,2,3,4,5,6,27,26,28])
vacinas = vacinas.rename(columns={'Hepatite A (HA)':'A', 'Hepatite B (HB)':'B'})
vacinas
```

```python
vacinas = pd.melt(vacinas, 'Ano', var_name='Classe_viral', value_name='Vacinas')
vacinas
```

```python
vacinas.to_csv('../dados_limpos/vacinas_hep')
```
